```python
import pandas as pd
import xgboost as xgb
import numpy as np
from sklearn import datasets, linear_model
```

```python
X_train = pd.read_csv('X_train_prep')
```

```python
X_test = pd.read_csv('X_test_prep')
```

```python
X= pd.read_csv('/Users/mi0310/Desktop/JPMC/train.csv')
Y_train = X["Fare"]
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(X_train, Y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)
```

```python
y_pred
```

```python
np.savetxt("y_pred_updated.csv", y_pred, delimiter=",")
```

```python
xgdmat=xgb.DMatrix(X_train,Y_train)
our_params={'eta':0.1,'seed':0,'subsample':0.8,'colsample_bytree':0.8,'objective':'reg:linear','max_depth':3,'min_child_weight':1}
final_gb=xgb.train(our_params,xgdmat)
tesdmat=xgb.DMatrix(X_test)
y_pred=final_gb.predict(tesdmat)
print(y_pred)
```
